## Manipulating Data


```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing.isnull().sum()
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=False)
```

```python
car_sales_missing
```

```python
car_sales_missing.dropna(inplace = True)
# Inplace permet de mettre à jour
```

```python
car_sales_missing
```

```python
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
car_sales_missing_dropped.to_csv("car_sales_missing_dropped.csv")
```

```python
# Column from series
seats_column = pd.Series([5,5,5,5,5])
# New column called seats
car_sales['seats'] = seats_column
car_sales
```

```python
car_sales["seats"].fillna(5, inplace=True)
```

```python
car_sales
```

```python
# Column from Python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 100km"] = fuel_economy
car_sales
```

```python
 car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100km"]
```

```python
car_sales
```

```python
# Create a column from a single value
car_sales["number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales.dtypes
```

```python
car_sales
```

```python
car_sales = car_sales.drop("Total fuel used", axis=1)
# la colonne est sur l'axe 1
```
